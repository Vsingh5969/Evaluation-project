# VIF Techineque

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X=df.drop('Attrition',axis=1)
X
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
vif_data.sort_values(by='VIF',ascending=False)
```

we observe that some of the highly correlated (>0.7 or <-0.7) features are:

MonthlyIncome ==== TotalWorkingYears===JobLevel

YearsAtCompany===YearsInCurrentRole ===YearsWithCurrManager

PercentSalaryHike=== PerformanceRating

It is cause of concern here as we see the multicollinearity of two columns are more than 10.
 
 for good prediction i am droping both the columns along with the columns which shows 0 or no correlation, as keeping then is just increase the feature columns

# Spliting and Balancing the Dataset

```python
from sklearn.model_selection import train_test_split
#spliting target and feature
new_x=df.drop('Attrition',axis=1)
y=df['Attrition']

x_train,x_test,y_train,y_test=train_test_split(new_x,y,random_state=101,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler,minmax_scale
scale=minmax_scale(new_x)
```

```python
y_train.value_counts() # imblacing indicated
```

```python
179/923*100
```

```python
# balancing the data 
from imblearn.over_sampling import SMOTE
```

```python
sm=SMOTE(0.90)
x_n_train,y_n_train=sm.fit_resample(x_train,y_train)
```

```python
print('old y_train','\n',y_train.value_counts(),'\n')
print('new y_train','\n', y_n_train.value_counts())
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
Lr=LogisticRegression()
Lr.fit(x_n_train,y_n_train)
```
